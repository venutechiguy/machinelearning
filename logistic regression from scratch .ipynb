```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z)) # calculates the sigmoid function with the formula 1/1+e^-x
```

# sigmoid:
 in order to map the values of the probabilites, we use the sigmoid function.the function maps any real value in to the another value between 0 and 1 , In machine learning we use sigmoid to map predictions to probabilites.

```python
def logistic_loss(y, y_hat):
    return -np.mean(y*np.log(y_hat) + (1-y)*np.log(1-y_hat))# calculates the logistic loss 
                                                            # the formula for log loss = -ylog(y^)-(1-y)log(1-y^)
    # here y is the label must be 0 or 1
    # here y^ is the predicted value  given the set of features of x
```

# logistic loss function:
  formula: = -ylog(y^)-(1-y)log(1-y^)
  (x,y) belongs to the function 
  y is the labeled data  it must be the label belong to 0 or 1 
  y^ is the predicted value it predicted the value given the set of features of x

```python
def logistic_loss_grad(X, y, w, b):#calculates the gradient of the logistic loss function
    y_hat = sigmoid(np.dot(X, w) + b)#calculates the derivates of the sigmoid function witrh multiples of b
    dL_dw = np.dot(X.T, y_hat - y) / y.shape[0]## calculates the parameters w and b
    dL_db = np.mean(y_hat - y)
    return dL_dw, dL_db# returns the values of dw and dl
```

# logistic gradient 
Gradient descent changes the value of our weights in such a way that it always converges to minimum point or we can also say that, it aims at finding the optimal weights which minimize the loss function of our model.
